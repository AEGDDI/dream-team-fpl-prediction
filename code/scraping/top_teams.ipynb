```python
import os
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from getpass import getuser
from openpyxl import Workbook, load_workbook
```

```python

# Set up Selenium options
options = Options()
options.add_argument("--headless")  # Run Chrome in headless mode

# Set up Chrome driver
user = getuser()
webdriver_service = Service(r'C:\Users\{}\Downloads\chromedriver.exe'.format(user))
driver = webdriver.Chrome(service=webdriver_service, options=options)

# Load the webpage
url = "https://www.fantasyfootballpundit.com/fpl-dream-team/"
driver.get(url)
sleep(3)  # Allow time for the page to load dynamically

def parse_table(table, season):
    rows = table.find_elements(By.TAG_NAME, "tr")

    data = []
    for row in rows[1:]:
        columns = row.find_elements(By.TAG_NAME, "td")
        player = columns[0].text.strip()
        team = columns[1].text.strip()
        points = columns[2].text.strip()
        data.append([season, player, team, points])

    return data

def save_data_to_excel(data, filename):
    if os.path.exists(filename):
        workbook = load_workbook(filename)
        sheet = workbook.active
    else:
        workbook = Workbook()
        sheet = workbook.active
        # Add the column names to the first row
        column_names = ['Season', 'Player', 'Team', 'Points']
        sheet.append(column_names)
    
    for row in data:
        sheet.append(row)
    
    workbook.save(filename)
    print(f'Data successfully saved to {filename}')

# Create a new folder for data if it doesn't exist
folder_path = r"C:\Users\{}\Documents\GitHub\dream-team-fpl-prediction\data".format(user)
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Set the filename with the folder path
filename = os.path.join(folder_path, "dream_team.xlsx")

# Find all the tables on the webpage
tables = driver.find_elements(By.TAG_NAME, "table")

# Iterate over the tables and retrieve data
season_index = 1
for table in tables:
    season_element = table.find_element(By.XPATH, f'//*[@id="post-15166"]/div/h3[{season_index}]')
    season = season_element.text.strip()

    data = parse_table(table, season)
    save_data_to_excel(data, filename)

    season_index += 1

print(f'All data successfully saved to {filename}')

# Quit the browser
driver.quit()
```
